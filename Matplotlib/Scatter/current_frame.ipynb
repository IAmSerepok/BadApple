```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'frames/frame_3673.jpg' # <-- Path to image
image = Image.open(image_path)
pixel_array = np.array(image)
```

```python
height, width, channels = pixel_array.shape
```

```python
step = 20 # <-- This is sampling step value
dot_size = 50 # <-- This is dot size value
```

```python
height = height // step
width = width // step
```

```python
plt.subplots(figsize=(9, 7)) # image size

for i in range(width):

    for j in range(height):

        col = pixel_array[j * step, i * step]
        if np.any(col != 0):
            col = col / 255
            plt.scatter(i * step, height - j * step, color=col, s=dot_size)

plt.savefig('outp.jpeg', dpi=300)

plt.show()
```
