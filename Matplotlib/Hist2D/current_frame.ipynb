```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'frames/frame_355.jpg' # <-- Path to image
image = Image.open(image_path)
pixel_array = np.array(image)
```

```python
height, width, channels = pixel_array.shape
```

```python
step = 5 # <-- This is sampling step value
```

```python
height = height // step
width = width // step
```

```python
x_data = []
y_data = []
```

```python
for i in range(width):

    for j in range(height):

        col = pixel_array[j * step, i * step]

        v = 0
        for _1 in range(10, 250, 10):
            if col[0] < _1:
                v += 1

        for _ in range(v):
            x_data.append(i * step)
            y_data.append(pixel_array.shape[0] - j * step)
```

```python
my_bins = ([_1 * step for _1 in range(width)], [_2 * step for _2 in range(height)])
```

```python
plt.subplots(figsize=(11, 7)) # image size

plt.hist2d(x_data, y_data, bins=my_bins, cmap='Grays', vmax=25)

plt.colorbar()

plt.savefig('outp.jpeg', dpi=300)

plt.show()
```
